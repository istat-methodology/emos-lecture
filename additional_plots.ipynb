# Plots
This notebook generates all the plots in the paper.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import scienceplots

plt.style.use(['science', 'notebook'])
```

### Distribution of Active Business in Italy

```python
df_viz = pd.read_csv("data/active_enterprises_2024_it.csv")
df_sorted = df_viz.sort_values(["section", "division"]).reset_index(drop=True)

fig, ax = plt.subplots(figsize=(18, 8))

bars = ax.bar(
    df_sorted.index,
    df_sorted["n"],
    color="darkgrey"
)

current_idx = 0
ymax = df_sorted["n"].max()
for i, (parent, group) in enumerate(df_sorted.groupby("section", sort=False)):
    start = current_idx - 0.5
    end = current_idx + len(group) - 0.5

    if i % 2 == 0:
        ax.axvspan(start, end, facecolor="lightgrey", alpha=0.3, zorder=0)

    center = (start + end) / 2
    ax.text(
        center,
        ymax,
        parent,
        ha="center",
        va="bottom",
        fontsize=13,
        fontweight="bold",
        alpha=0.7,
        rotation=0
    )

    current_idx += len(group)

ax.set_xticks(df_sorted.index)
ax.set_xticklabels([f"{x:02}" for x in df_sorted["division"]], rotation=90)
ax.set_xlabel("ATECO Division")
ax.set_ylabel("Count")
ax.set_title("Active Businesses in Italy by ATECO Division (December 2024)")

ax.margins(x=0)
ax.set_xlim(-0.9, len(df_sorted)-0.1)

formatter = FuncFormatter(lambda x, _: f'{int(x/1000)}k' if x >= 1000 else str(int(x)))
ax.yaxis.set_major_formatter(formatter)

plt.grid(True, which="major", axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
